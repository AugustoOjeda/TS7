### Introducción teórica


La transformada Z es una herramienta fundamental en el analisis y diseños de sistemas digitales, ya que permite representar señales y sistemas discretos en el dominio complejo, Es analoga a la transformada de Laplace en sistemas continuos, pero adaptada a secuencias discretas en el tiempo. Para una secuencia x[n], la transformada Z se define como:
$$X(z)=\sum_{k=-\infty}^{\infty}x[k]re^{-j \omega_0 n} $$
Para avanzar con esta explicacion teorica vamos a definir $z=-j \omega_0 n$ Esta representación permite estudiar las propiedades de estabilidad y causalidad de los sistemas ya que los polos y ceros de T(z) determinan el comportamiento del sistema. En general lo que se busca son series convergentes es decir series dondes los polos se encuentran en el semiplano izquierdo o que se acumulen en el 0. El valor r me permite barrer todo el plano complejo y genera periodicidad al tener un dominio circular, ahora bien no todo r genera convergencia y me permite analizar tanto estados transitarios como los estados permanentes. 

La mayoria de los sistemas pueden expresarse como una razon de dos polinomios
$$T(z)=\frac{N(z)}{D(z)}=\frac{\sum_{k=0}^{M}b_kz^{-k}}{\sum_{k=0}^{M}a_kz^{-k}}$$
Los ceros son los valores de z que anulan N(z), mientras que los polos son los que anulan D(z). La ubicación de estos polos y ceros en el plano Z define el comportamiento del sistema en frecuencia y su estabilidad.La respuesta en frecuencia de un sistea discreto se obtiene al evaluar T(z) sobre el circulo unitario del plano complejo $$T(e^{-j \omega})=H(z)|_{z=e^{-j \omega}}$$

Esta evaluacion relaciona directamente la transformada Z con la DTFT ya que describe como el sistema modifica la amplitud y la fase de cada componente de frecuencia de la señal de entrada. El modulo indicla la ganancia en magnitud mientras que su argumento el desfase introducido por el sistema.

Para ello, se utilizó la función freqz de la librería Signal de SciPy. Esta función permite al usuario definir cuatro parámetros principales: los coeficientes $b_k$ y $a_k$, junto con las opciones de frecuencia. En este caso, se adoptó $a_0=1$ y se consideraron nulos los demás coeficientes $a_k$ (para k>0), motivo por el cual se introdujo simplemente a=1. En cambio, los valores de $b_k$ se asignaron de acuerdo con los obtenidos a partir de las ecuaciones teóricas.

Finalmente, se fijó una frecuencia de muestreo arbitraria de fs=1000 Hz, con el propósito de expresar los resultados directamente en unidades de frecuencia (Hz) en lugar de radianes por muestra (rad/muestra), evitando así la necesidad de un reescalado posterior. Para utilizar el codigo se definieron los siguientes vectores y cte

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

#Coeficientes
ba=[1,1,1,1]
bb=[1,1,1,1,1]
bc=[1,-1]
bd=[1,-1]

#cte 
fs=1000 #Hz
```

```python
#freqz para cada caso
wa,ha=sig.freqz(b=ba,a=1,whole=False,fs=fs) 
wb,hb=sig.freqz(b=bb,a=1,whole=False, fs=fs)
wc,hc=sig.freqz(b=bc,a=1,whole=False, fs=fs)
wd,hd=sig.freqz(b=bd,a=1,whole=False, fs=fs)
```

Al utilizar freqz obtenemos dos salidas, las cual en generar se representan como w y h, en este caso se dio una señalizacion a cada salida dependiendo del ejercicio donde w equivale a las frecuencias en las que se calculó h, expresadas en las mismas unidades que la frecuencia de muestreo fs y h equivale a la respuesta en frecuencia, representada mediante números complejos (contiene tanto la magnitud como la fase del sistema). Con esto en mente se calculo el modulo y la fase de las siguiente manera:

```python
#maginitud y fase
maga=np.abs(ha) #Modulo T(e^jΩ)
fasea=np.unwrap(np.angle(ha)) #Fase T(e^jΩ) (radianes)
magb=np.abs(hb) #Modulo T(e^jΩ)
faseb=np.unwrap(np.angle(hb)) #Fase T(e^jΩ) (radianes) 
magc=np.abs(hc) #Modulo T(e^jΩ)
fasec=np.unwrap(np.angle(hc)) #Fase T(e^jΩ) (radianes)
magd=np.abs(hd) #Modulo T(e^jΩ)
fased=np.unwrap(np.angle(hd)) #Fase T(e^jΩ) (radianes)
```

```python
plt.figure()
plt.plot(wa, maga)
plt.title('Respuesta en frecuencia |t(e^{jΩ})|, para el ejercicio a')
plt.ylabel('Modulo |t(e^{jΩ})|')
plt.xlabel('frecuencia (Hz)')
```

```python
plt.figure()
plt.plot(wa, fasea)
plt.title('Respuesta en frecuencia de la fase para el ejercicio a')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Fase [rad]')
```
